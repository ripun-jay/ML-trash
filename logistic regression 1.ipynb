```python
# import required packages

from sklearn import datasets
from sklearn.linear_model import LogisticRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading Data

iris = datasets.load_iris()
print(iris.keys())
```

```python
# View data, target, DESCR for domain knowledge

iris.data
```

```python
# Defining Feature(X) and Labels(Y or y)

# here taking only one feature into picture for understing with plot
features = iris.data[:,3:]

# here we taking labels is binary form for being iris veriginc class or not
labels = (iris.target == 2).astype(int)
```

```python
# View Labels and features
print(f" Features are: \n{features}\n\n\n and labels are:\n {labels}")
```

```python
# View shape of labels and Features to validate that does they capble of internal multiplication

print(f"feature --> {np.shape(features)}\nlabel --> {np.shape(labels)}")
```

```python
# select an model

clf = LogisticRegression()
```

```python
# Fit the data into model - TRAINING

clf.fit(features, labels)
```

```python
# Pridict some Random value class that is it iris verginac or not -->  0 or 1

predicted = clf.predict([[1.6]])
predicted
```

```python
# Predicting multiple existing value for idea about error

clf.predict([[0.2],[1.8],[1.7],[1.4],[1.5]])
```

```python
# Determining How much any point is verginica

clf.predict_proba([[0.2],[1.8],[1.7],[1.4],[1.5]])
```

```python
# do ploting to undertand sigma function

X_plt = np.linspace(-4,4,1000).reshape(-1,1)
print(X_plt[:10])
```

```python
Y_plt = clf.predict_proba(X_plt)[:,1]
Y_plt
```

```python
plt.plot(X_plt, Y_plt, "r-", alpha= 0.8)
plt.show()
```
